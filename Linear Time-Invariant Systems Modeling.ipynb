## Linear Time-Invariant Systems Modeling

### Description

This Jupyter notebook project aims to model a simple linear time-invariant (LTI) system. Linear systems are fundamental in signal processing, control theory, and various engineering applications. In this project, we will use Python and Jupyter notebooks to demonstrate the principles of LTI systems.


### Mathematical Background

A linear time-invariant system can be represented by the convolution integral:

 $y(t) = \int_{-\infty}^{\infty} x(\tau)h(t-\tau) d\tau $

Where:
- $ y(t) $ is the output signal.
- $ x(t) $ is the input signal.
- $ h(t) $ is the system's impulse response.


````python
"""
Generate Convolution Plot

This Python code generates a plot illustrating the convolution of an input signal x(t) with an impulse response h(t).

The code first defines time values and creates example signals:
- Input signal x(t): A Gaussian pulse.
- Impulse response h(t): A sinc function.

Then, it performs convolution between x(t) and h(t) and normalizes the result.

The code creates three subplots:
1. Input Signal x(t): Shows the Gaussian pulse.
2. Impulse Response h(t): Shows the sinc function.
3. Output Signal y(t) (Convolution): Shows the result of the convolution.

The plots are stylized with LaTeX-like notations for labels and titles.

This code is intended for educational purposes to demonstrate the concept of convolution in signal processing.
"""

import numpy as np
import matplotlib.pyplot as plt

# Define time values
t = np.linspace(-5, 5, 1000)

# Define input signal x(t) and impulse response h(t)
x = np.exp(-t**2)  # Example input signal (Gaussian)
h = np.sinc(t)     # Example impulse response (Sinc)

# Perform convolution
y = np.convolve(x, h, 'same') / sum(h)  # Convolution and normalization

# Create subplots
plt.figure(figsize=(12, 4))

# Subplot 1: Input Signal x(t)
plt.subplot(131)
plt.plot(t, x, 'b', label='$x(t)$')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Input Signal $x(t)$')
plt.legend()

# Subplot 2: Impulse Response h(t)
plt.subplot(132)
plt.plot(t, h, 'g', label='$h(t)$')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Impulse Response $h(t)$')
plt.legend()

# Subplot 3: Output Signal y(t) (Convolution)
plt.subplot(133)
plt.plot(t, y, 'r', label='$y(t)$')
plt.xlabel('Time $t$')
plt.ylabel('Amplitude')
plt.title('Output Signal $y(t)$ (Convolution)')
plt.legend()

# Adjust layout and show plot
plt.tight_layout()
plt.show()
````
